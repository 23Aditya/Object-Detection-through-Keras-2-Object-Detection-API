We have to now first of all create a label map file (having .pbtxt extention) which will be having mappings from different class label names to the integer categories. Tensorflow 2 Object Detection API by default considers Background category as belonging to integer class label 0 and it's not included in this file.

In order to create a .pbtxt label map file, we have to first determine that how many categories are there in our training data and the best way to determine this is from annotation file (usually a .csv file but sometimes, we don't directly get a .csv file but instead we get and we have extract what is called a .xml file and from this .xml file you have to convert it into .csv file)

```python
# we r creating a function ---spliting function  whick makes groups of images(283)

def split_data_by_filename(annotations,column_name):
  data = namedtuple("data",["filename","object"])
  imggroup = annotations.groupby(column_name)
  return [data(filename,imggroup.get_group(x)) for filename,x in zip(imggroup.groups.keys(),imggroup.groups)] # we r doing list compresshion
```

```python
# this fun. will create tenserflow examples
# and call 283 tyms of every grp...

def create_tf_example(imggroup,path):
  with tf.gfile.GFile(os.path.join(path, '{}'.format(imggroup.filename)), 'rb') as fid:  # here we r reading file in binarry format
        encoded_jpg = fid.read()
  encoded_jpg_io = io.BytesIO(encoded_jpg)     # image pixels encode into bytes
  image = Image.open(encoded_jpg_io)
  width, height = image.size

  filename = imggroup.filename.encode('utf8')   # we converted our string fine name into utf8  
  image_format = b'jpg'
  xmins = []                                      # and  these details are compulsory...
  xmaxs = []
  ymins = []
  ymaxs = []
  classes_text = []
  classes = []                                   # from here

  for index, row in imggroup.object.iterrows():                                        #
      xmins.append(row['xmin'] / width)
      xmaxs.append(row['xmax'] / width)
      ymins.append(row['ymin'] / height)
      ymaxs.append(row['ymax'] / height)
      classes_text.append(label_map_dict[row['gt_labels']].encode('utf8'))
      classes.append(row['gt_labels'])

  tf_example = tf.train.Example(features=tf.train.Features(feature={
        'image/height': dataset_util.int64_feature(height),
        'image/width': dataset_util.int64_feature(width),                                        #  These info. are mandatory 
        'image/filename': dataset_util.bytes_feature(filename),
        'image/source_id': dataset_util.bytes_feature(filename),
        'image/encoded': dataset_util.bytes_feature(encoded_jpg),
        'image/format': dataset_util.bytes_feature(image_format),
        'image/object/bbox/xmin': dataset_util.float_list_feature(xmins),
        'image/object/bbox/xmax': dataset_util.float_list_feature(xmaxs),
        'image/object/bbox/ymin': dataset_util.float_list_feature(ymins),
        'image/object/bbox/ymax': dataset_util.float_list_feature(ymaxs),
        'image/object/class/text': dataset_util.bytes_list_feature(classes_text),
        'image/object/class/label': dataset_util.int64_list_feature(classes),        # 
    }))
  return tf_example
```

```python
out_training_img = {
        'image/height': tf.io.FixedLenFeature(shape=[],dtype=tf.int64),
        'image/width': tf.io.FixedLenFeature(shape=[],dtype=tf.int64),
        'image/filename': tf.io.FixedLenFeature(shape=[],dtype=tf.string),
        'image/source_id': tf.io.FixedLenFeature(shape=[],dtype=tf.string),
        'image/encoded': tf.io.FixedLenFeature(shape=[],dtype=tf.string),
        'image/format': tf.io.FixedLenFeature(shape=[],dtype=tf.string),
        'image/object/bbox/xmin': tf.io.VarLenFeature(dtype=tf.float32),
        'image/object/bbox/xmax': tf.io.VarLenFeature(dtype=tf.float32),
        'image/object/bbox/ymin': tf.io.VarLenFeature(dtype=tf.float32),
        'image/object/bbox/ymax': tf.io.VarLenFeature(dtype=tf.float32),
        'image/object/class/text': tf.io.FixedLenSequenceFeature(shape=[],dtype=tf.string,allow_missing=True),
        'image/object/class/label': tf.io.VarLenFeature(dtype=tf.int64),
    }
```

```python
# read a single image
def parse_image(example_proto):
  return tf.io.parse_single_example(serialized=example_proto,features=out_training_img)
```

```python
# mediatory lib. for infrence...
import random
from io import BytesIO
import matplotlib.pyplot as plt
import numpy as np
import tensorflow as tf
from object_detection.utils import ops
from object_detection.utils import visualization_utils as viz
from object_detection.utils.label_map_util import create_category_index_from_labelmap
```

```python
def load_test_image(path):
  test_image_data = tf.io.gfile.GFile(path,'rb').read()
  test_image = Image.open(BytesIO(test_image_data))
  width,height = test_image.size
  shape = (width,height,3)
  test_image = np.array(test_image.getdata())
  test_image = test_image.reshape(shape).astype("uint8")
  return test_image
```

```python
def make_inference(trained_model,image):
  image = np.asarray(image)
  image_tensor = tf.convert_to_tensor(image)
  image_tensor = image_tensor[tf.newaxis,...]
  image_predictions = trained_model(image_tensor)
  
  num_detections = int(image_predictions.pop("num_detections"))
  image_predictions = {key:value[0,:num_detections].numpy() for key,value in image_predictions.items()}
  image_predictions["num_detections"] = num_detections
  image_predictions["detection_classes"] = image_predictions["detection_classes"].astype("int64")

  if "detection_masks" in image_predictions:
    detection_masks_reframed = ops.reframe_box_masks_to_image_masks(image_predictions["detection_masks"],
                                                                    image_predictions["detection_boxes"],image.shape[0],
                                                                    image.shape[1])
    detection_masks_reframed = tf.cast(detection_masks_reframed > 0.5,tf.unint8)
    image_predictions["detection_masks_reframed"] = detection_masks_reframed.numpy()
  return image_predictions
```

```python
labels_path = "/content/drive/MyDrive/tf2odapi/workspace/shelf_images_training_demo/annotations/grocerydataset/label_map.pbtxt"
category_idx = create_category_index_from_labelmap(labels_path,use_display_name=True)
model_path = "/content/drive/MyDrive/tf2odapi/workspace/shelf_images_training_demo/saved_model"
trained_model = tf.saved_model.load(model_path)
test_images = list(glob.glob("/content/drive/MyDrive/ShelfImages/test/*.JPG"))
random.shuffle(test_images)
test_images = test_images[:16]
```

```python
for image_path in test_images:

  image = load_test_image(image_path)
  image_predictions = make_inference(trained_model,image)
  detection_masks = image_predictions.get("detection_masks_reframed",None)
  viz.visualize_boxes_and_labels_on_image_array(image,image_predictions["detection_boxes"],image_predictions["detection_classes"],
                                                image_predictions["detection_scores"],category_idx,instance_masks=detection_masks,
                                                use_normalized_coordinates=True,line_thickness=5)
  plt.figure(figsize=(24,24))
  plt.imshow(image)
  plt.savefig(f'detections_{image_path.split("/")[-1]}')
```
